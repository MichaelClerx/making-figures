# General tip 5: Show your workings

Resist the temptation to hide ugly data sets or skip "obvious" intermediate steps.

## Examples

### Hiding negative results

Let's say you've measured x and y, and expected a strong correlation.
You've tried a few obvious ways of plotting and analysing, but still can't see the expected result.
You write something like

> We measured x and y, but no found no correlation (data not shown).

and continue to a different measurement or a more complex analysis.

But why tell the reader that an expected thing didn't happen (and expect them to believe you), when you can easily show it?

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.uniform(1, 10, 100)
y = np.random.uniform(-3, 3, 100)

fig = plt.figure(figsize=(4, 3))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel('x (m)')
ax.set_ylabel('y (V)')
ax.set_xlim(0, 11)
ax.set_ylim(-3.5, 3.5)
ax.plot(x, y, 'o')
plt.show()
```

### Skipping ahead

In this example, we've measured two things and discovered that there is a huge size difference.
So much so, that plotting them together just looks silly, so we adjust the y scale to get a sensible plot:

```python
h_e = np.random.normal(3, 0.3, 100)
h_a = np.random.normal(2, 0.2, 100)

fig = plt.figure(figsize=(4, 3))
fig.subplots_adjust(wspace=0.7)

ax = fig.add_subplot(1, 2, 1)
ax.set_ylabel('Size (m)')
ax.boxplot(h_e, labels=['Elephants'])

ax = fig.add_subplot(1, 2, 2)
ax.set_ylabel('Size (mm)')
ax.boxplot(h_a, labels=['Ants'])

plt.show()
```

_Figure 1: Ants are smaller than elephants (note the different y-axes)._

Nothing wrong with this figure. Entirely correct and shows all the details. But does it get the message across that elephants are bigger?

Why not let the reader make the same steps as you did, and show them the raw data first:

```python
fig = plt.figure(figsize=(6, 3))
fig.subplots_adjust(wspace=1.5)
grid = fig.add_gridspec(1, 4)

ax = fig.add_subplot(grid[0, 0:2])
ax.set_ylabel('Size (m)')
ax.boxplot([h_e, h_a * 1e-3], labels=['Elephants', 'Ants'])

ax = fig.add_subplot(grid[0, 2])
ax.set_ylabel('Size (m)')
ax.boxplot(h_e, labels=['Elephants'])

ax = fig.add_subplot(grid[0, 3])
ax.set_ylabel('Size (mm)')
ax.boxplot(h_a, labels=['Ants'])

plt.show()
```

This is a contrived example, but it illustrates a general pattern:

1. Statement saying "obvious analysis didn't work".
2. Section presenting new and more complex analysis method.
3. Result.

If the new analysis is complicated, or if there is a more obvious or established method, readers will be left feeling uneasy.
Replace the statement in step 1 with a graph to let the reader check your workings and motivate the new analysis.
