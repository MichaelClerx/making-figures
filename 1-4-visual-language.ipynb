# General tip 4: Use visual language

### 4.1 The most important part of the figure should be the most eye-catching.

A good way to achieve this is to start by making everything as boring and uniform as possible, and then to add accents _deliberately_.

People will read meaning into colours, sizes, shapes, proximity, fonts... Try not to "say" anything unless you mean it.

For example:
- Organise your figure on a regular grid.
  - Use size (bigger) or placement (central or at the top) to indicate importance
  - Place related things close together. Group things in boxes if needed. Draw hierarchical relationships (e.g. containments).
- Aim for the same level of detail in all parts of your figure. E.g. if you are studying cells and draw them as circles, don't add a detailed microscope.
  - Use increased detail to indicate importantance and draw the eye.
- Remove unnecessary differences in fonts, font sizes, line widths.
  - Indicate importance with **weight** or a bigger font, but keep it to a minimum.
- Be consistent with colours, even from figure to figure.
  - E.g. use one colour for voltage, another for current.
  - Or use one colour for cell 1, another for cell 2, etc., so that people can connect data from different figures.


### 4.2 Consciously use "visual communication".

The wikipedia page on [visual communication](https://en.wikipedia.org/wiki/Visual_communication) lists "seven components of visual communication".

#### Hierarchy, balance, and figure-ground are useful for your overal layout.

"**Hierarchy**" means laying out the figure so that readers can read it piece by piece.
This means e.g. grouping things in boxes, putting important things at the top, showing relationships between items (e.g. showing that X contains Y by putting the drawing of Y inside the drawing of X), adding arrows and labels, and generally guiding the viewer through the figure.

"**Balance**" relates to symmetry, which can mean spatial symmetry but also equal "weighting" of components by giving them equally bold colours, equally strong lines, etc. Purposeful imbalance can be a useful tool to draw the eye. (A lot of 4.1 can be rephrased as: start with a balanced figure, then add imbalance to draw focus to the important parts).

"**Figure-ground**" means background vs foreground, where foreground is the thing you see (e.g. the letters and lines on a page), and background is the thing you ignore (e.g. the white page itself).
Scientific figures usually have little or no content in the background.

#### Colour, shape and tone are useful to make distinctions.

Examples are using different **colours** for different lines in a graph, using differently **shaped** markers in scatter plots, and using darker and lighter "**tones**" of colours to make a distinction or to indicate a grouping.
**Size** can be used for similar purposes.

Colour can be very effective and is often used as the primary tool to make distinctions, but keep in mind that [not everybody can easily distinguish between different hues](https://en.wikipedia.org/wiki/Color_blindness).
In particular, if you only need 2 colours never pick red and green!

#### "Texture" is not as frequently used.

The 7th component, visual **texture**, refers to "the illusion of having physical texture" (see [here](https://en.wikipedia.org/wiki/Texture_(visual_arts))).
Although not as common in scientific figures, examples of texture are [cross-hatching](https://en.wikipedia.org/wiki/Hatching) or the inclusion of fine detail in cartoons to give a more realistic appearance.

### 4.3 Avoid unintended messages.

Try to avoid unintended messages or distracting details.
For example, if you are drawing a diagram and need to represent the concept of an ECG, don't use something like the figure below (adapted from [this public domain image](https://commons.wikimedia.org/wiki/File:12leadECG.jpg)):

![ecg](./figures-1/ecg-1.jpg)

Any cardiologists in the room might stop listening and start diagnosing.
(This happened to a friend and me and ate up 20 minutes of our presentation time)

![ecg-cartoon](./figures-1/ecg-2.svg)

Cartoons like the one above can help in these situations.
But even then, try to avoid adding details unless you're very sure of them.

If using cartoons of humans, be careful not to propagate stereotypes.

## Examples

![myokit](./figures-1/overview-v1.png)

An early version of a diagram describing [myokit](https://doi.org/10.1016/j.pbiomolbio.2015.12.008). **There are lots of things wrong with this figure.**

Hierarchy is missing: there is no clear focal point, so it's unclear where we should start viewing this figure. Should we be going top to bottom? Left to right?

There's no clear structure to the layout. Almost every item is at its own width or height, giving a messy appearance that's quite distracting. (It's fine to put things in weird places, as long as you mean something by it. In this case I did not.)

Spatially, the layout isn't balanced. And the same goes for colour, tone and level of detail: the cell model is simply a full figure scaled down to fit, and the journal icon includes a high-res confocal microscopy image. The model, journal, and 2D simulation result are in full colour. This unintentionally draws the eye towards those items, and distracts from the (very boring) "Myokit" box (which is hidden away in a spot to the right & below the figure's center).

For no particular reason, the scripts on the left are a nice blue: We have taken one of our most powerful tools, colour, and flung it out the window.

![myokit](./figures-1/overview-v7.png)

Seven iterations later, this improved (but far from perfect) version was used in the publication.

The layout is now much more balanced, and the (loose) use of a grid has removed a lot of distraction. The relationship between Myokit and its components is indicated visually: the "Myokit" box literally contains the other boxes, while dependencies are linked from outside it. Arrows indicate a workflow from model and user to end result. Colour is used very sparingly, which makes this suitable for black-and-white printing, but also makes the Myokit logo stand out.

The stupid confocal image is still present, however, and the 2D simulation result is very large and a lot darker than the rest (tone imbalance), causing it to draw an unwarranted amount of attention (Myokit is primarily a single-cell simulation tool). The CellML repository icon also stands out by being a bit awkward (why those very faint lines? Why that bold font? Why the vertically centered text?).

Finally, I was helpfully & embarassingly reminded that scientists are not all balding men with glasses, and so the version in my thesis was updated as below:

This was perhaps a step in the right direction, but it may have been better still to use e.g. an outline of a generic human being (see for example the [unicef](https://en.wikipedia.org/wiki/UNICEF) or [physiome project](https://physiomeproject.org/) logos).

### Balance

![img](./figures-1/balance-1.svg)

An example of a balanced layout.
Boxes are laid out symmetrically.
Each occupies the same amount of space and no single box has a more prominent position than the others.
Although colour is used to distinguish between items, the "tone" of each box is roughly the same, so that no particular box steals focus.
This does the job of conveying that all four parts are equally important, but it's also incredibly boring.

![img](./figures-1/balance-2.svg)

A very unbalanced layout. The space is not filled evenly; there's no symmetry; colors, shapes and level of detail are all "weighted" towards the "THIS" item.

If done well, lack of balance can create tension that draws the viewer in.
Done without skill (see above), it can be very ugly too.

### Making distinction with colour, shape, tone

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 14, 100)
y1 = 1 + 0.2 * x + 0.4 * x**2
y2 = 6 + 0.4 * x + 0.3 * x**2
y3 = 12 + 0.2 * x + 0.2 * x**2
y4 = 18 + 0.4 * x + 0.1 * x**2

plt.figure(figsize=(8, 5))
plt.xlabel('Time (hours)')
plt.ylabel('Amount')
plt.plot(x, y1, label='Thing 1')
plt.plot(x, y2, label='Thing 2')
plt.plot(x, y3, label='Thing 3')
plt.plot(x, y4, label='Thing 4')
plt.legend(loc='upper left')
plt.show()
```

The example above shows the amounts of four different things, increasing over time.
We've used matplotlib's default settings, and so the four lines were assigned colours from the ["tableau 10" set](https://matplotlib.org/stable/gallery/color/named_colors.html).

An extended ["tab20"](https://matplotlib.org/stable/gallery/color/colormap_reference.html) set is available which repeats all the colours in two different intensities (tones).
We can use this to add a second dimension to our visualisation.

```python
import matplotlib
cmap = matplotlib.cm.get_cmap('tab20')

x = np.linspace(0, 14, 100)
y1b = 1 + 0.18 * x + 0.38 * x**2
y2b = 6 + 0.38 * x + 0.29 * x**2
y3b = 12 + 0.18 * x + 0.19 * x**2
y4b = 18 + 0.38 * x + 0.07 * x**2

plt.figure(figsize=(8, 5))
plt.xlabel('Time (hours)')
plt.ylabel('Amount')
plt.plot(x, y1, color=cmap(0), label='Thing 1 (fast)')
plt.plot(x, y1b, color=cmap(1), label='Thing 1 (slow)')
plt.plot(x, y2, color=cmap(2), label='Thing 2 (fast)')
plt.plot(x, y2b, color=cmap(3), label='Thing 2 (slow)')
plt.plot(x, y3, color=cmap(4), label='Thing 3 (fast)')
plt.plot(x, y3b, color=cmap(5), label='Thing 3 (slow)')
plt.plot(x, y4, color=cmap(6), label='Thing 4 (fast)')
plt.plot(x, y4b, color=cmap(7), label='Thing 4 (slow)')

plt.legend(loc='upper left')
plt.show()
```

If we expect our graph to be plotted in black and white, we might want to consider a different way of distinguishing the lines:

```python
plt.figure(figsize=(8, 5))
plt.xlabel('Time (hours)')
plt.ylabel('Amount')
plt.plot(x, y1, ls='-', color='black', label='Thing 1')
plt.plot(x, y2, ls=':', color='black', label='Thing 2')
plt.plot(x, y3, ls='--', color='black', label='Thing 3')
plt.plot(x, y4, ls='-.', color='black', label='Thing 4')
plt.legend(loc='upper left')
plt.show()
```

Note that this doesn't quite given them all the same weight. The solid line stands out a bit more than the others. Use this to your advantage if you want to highlight e.g. a baseline or gold-standard measurement, or the performance of a new method compared to its predecessors.

As before, we can play with tone to add a second dimension (dark = fast, light = slow):

```python
plt.figure(figsize=(8, 5))
plt.xlabel('Time (hours)')
plt.ylabel('Amount')
plt.plot(x, y1, ls='-', color='black', label='Thing 1 (fast)')
plt.plot(x, y1b, ls='-', color='grey', label='Thing 1 (slow)')
plt.plot(x, y2, ls=':', color='black', label='Thing 2 (fast)')
plt.plot(x, y2b, ls=':', color='grey', label='Thing 2 (slow)')
plt.plot(x, y3, ls='--', color='black', label='Thing 3 (fast)')
plt.plot(x, y3b, ls='--', color='grey', label='Thing 3 (slow)')
plt.plot(x, y4, ls='-.', color='black', label='Thing 4 (fast)')
plt.plot(x, y4b, ls='-.', color='grey', label='Thing 4 (slow)')
plt.legend(loc='upper left')
plt.show()
```

But we can also consider other methods, e.g. using line width:

```python
plt.figure(figsize=(8, 5))
plt.xlabel('Time (hours)')
plt.ylabel('Amount')
w1 = 2
w2 = 1
plt.plot(x, y1, ls='-', lw=w1, color='black', label='Thing 1 (fast)')
plt.plot(x, y1b, ls='-', lw=w2, color='black', label='Thing 1 (slow)')
plt.plot(x, y2, ls=':', lw=w1, color='black', label='Thing 2 (fast)')
plt.plot(x, y2b, ls=':', lw=w2, color='black', label='Thing 2 (slow)')
plt.plot(x, y3, ls='--', lw=w1, color='black', label='Thing 3 (fast)')
plt.plot(x, y3b, ls='--', lw=w2, color='black', label='Thing 3 (slow)')
plt.plot(x, y4, ls='-.', lw=w1, color='black', label='Thing 4 (fast)')
plt.plot(x, y4b, ls='-.', lw=w2, color='black', label='Thing 4 (slow)')
plt.legend(loc='upper left')
plt.show()
```

Finally, an example using [shapes](https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html):

```python
# Downsample the data
k = 10
sx = x[::k]
sy1, sy2, sy3, sy4 = y1[::k], y2[::k], y3[::k], y4[::k]
sy1b, sy2b, sy3b, sy4b = y1b[::k], y2b[::k], y3b[::k], y4b[::k]

ms = 7
plt.figure(figsize=(8, 5))
plt.xlabel('Time (hours)')
plt.ylabel('Amount')
plt.plot(sx, sy1, marker='^', markersize=ms, fillstyle='full', color='black', label='Thing 1 (fast)')
plt.plot(sx, sy1b, marker='^', markersize=ms, fillstyle='none', color='grey', label='Thing 1 (slow)')
plt.plot(sx, sy2, marker='s', markersize=ms, fillstyle='full', color='black', label='Thing 2 (fast)')
plt.plot(sx, sy2b, marker='s', markersize=ms, fillstyle='none', color='grey', label='Thing 2 (slow)')
plt.plot(sx, sy3, marker='o', markersize=ms, fillstyle='full', color='black', label='Thing 3 (fast)')
plt.plot(sx, sy3b, marker='o', markersize=ms, fillstyle='none', color='grey', label='Thing 3 (slow)')
plt.plot(sx, sy4, marker='v', markersize=ms, fillstyle='full', color='black', label='Thing 4 (fast)')
plt.plot(sx, sy4b, marker='v', markersize=ms, fillstyle='none', color='grey', label='Thing 4 (slow)')
plt.legend(loc='upper left')
plt.show()
```

Notice how we quite quickly run out of shapes that don't look a lot like circles, especially when marker sizes are small.

```python
ms = 6
plt.figure(figsize=(8, 5))
plt.xlabel('Time (hours)')
plt.ylabel('Amount')
plt.plot(sx, sy1, marker='o', markersize=ms, fillstyle='full', color='black', label='Thing 1 (fast)')
plt.plot(sx, sy2, marker='d', markersize=ms, fillstyle='full', color='black', label='Thing 2 (fast)')
plt.plot(sx, sy3, marker='p', markersize=ms, fillstyle='full', color='black', label='Thing 3 (fast)')
plt.plot(sx, sy4, marker='h', markersize=ms, fillstyle='full', color='black', label='Thing 4 (fast)')
plt.legend(loc='upper left')
plt.show()
```

Avoid this kind of thing!

### Small multiples

Another way to differentiate is to simply make separate graphs, known as [small multiples](https://en.wikipedia.org/wiki/Small_multiple).

```python
plt.figure(figsize=(16, 4))
ylim = 0, 72

ax = plt.subplot(1, 4, 1)
ax.set_xlabel('Time (hours)')
ax.set_ylabel('Amount')
ax.set_ylim(*ylim)
ax.plot(sx, sy1, label='Thing 1 (fast)')
ax.plot(sx, sy1b,  label='Thing 1 (slow)')
ax.legend(loc='upper left')

ax = plt.subplot(1, 4, 2)
ax.set_xlabel('Time (hours)')
ax.set_ylim(*ylim)
ax.plot(sx, sy2, label='Thing 2 (fast)')
ax.plot(sx, sy2b, label='Thing 2 (slow)')
ax.legend(loc='upper left')

ax = plt.subplot(1, 4, 3)
ax.set_xlabel('Time (hours)')
ax.set_ylim(*ylim)
ax.plot(sx, sy3, label='Thing 3 (fast)')
ax.plot(sx, sy3b, label='Thing 3 (slow)')
ax.legend(loc='upper left')

ax = plt.subplot(1, 4, 4)
ax.set_xlabel('Time (hours)')
ax.set_ylim(*ylim)
ax.plot(sx, sy4, label='Thing 4 (fast)')
ax.plot(sx, sy4b, label='Thing 4 (slow)')
ax.legend(loc='upper left')

plt.show()
```
